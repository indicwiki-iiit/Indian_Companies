```python
temp=[]
for i in range(len(data)):
    print(data['CIN'][i],i)
    inputElement = driver.find_element_by_id('searchid')
    if type(i)!=float:
        try:
            inputElement.send_keys(data['CIN'][i])
            inputElement.send_keys(Keys.ENTER)
            ele=driver.find_elements(By.XPATH,value='//*[@id="block-system-main"]/div[2]/div[1]/div[1]/div[1]/table/tbody/tr[8]/td[2]/p')
            x=ele[0].text
            temp.append(x)
            print(x)
        except:
            temp.append(np.nan)
    else:
        temp.append(np.nan)
        continue
```

```python
data['date_of_incorp']=temp
```
