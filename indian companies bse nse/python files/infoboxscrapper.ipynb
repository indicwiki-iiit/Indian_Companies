```python
import pandas as pd
import numpy as np
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.chrome.service import Service
import openpyxl
```

```python
s=Service("D:\chromedrive\chromedriver.exe")
driver=webdriver.Chrome(service=s)
```

```python
data=pd.read_csv("fullwikicmp - fullwikicmp (1).csv")
```

```python
data['url']
```

```python
li=[]
for i,j in zip(data['Company'],data['url']):
    try:
        driver.get(j)
        table1 = driver.find_elements(by=By.XPATH,value = "//table[@class='infobox vcard']/tbody/tr/th")
        table2 = driver.find_elements(by=By.XPATH,value = "//table[@class='infobox vcard']/tbody/tr/td")
        t = dict()
        t['Company'] = i
        for k,l in zip(table1,table2[1:]):
            t[k.text] = l.text
        li.append(t)
        print(i,j)
    except:
        li.append(np.nan)
```

```python
len(li)
```

```python
d=pd.DataFrame(li)
```
