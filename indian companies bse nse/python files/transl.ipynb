```python
l1=[]
l2=[]
l3=[]
l4=[]
l5=[]
p=0
for i in x['Description']:
    if type(i)!=float:
        k=i.split("\n\n")
        if len(k)==5:
            l1.append(k[0])
            l2.append(k[1])
            l3.append(k[2])
            l4.append(k[3])
            l5.append(k[4])
        elif len(k)==4:
            l1.append(k[0])
            l2.append(k[1])
            l3.append(np.nan)
            l4.append(k[2])
            l5.append(k[3])
    else:
        l1.append(np.nan)
        l2.append(np.nan)
        l3.append(np.nan)
        l4.append(np.nan)
        l5.append(np.nan)
    p+=1
    print(p,len(l1))
    
```

```python
p
```

```python
x['Description_1']=l1
x['Description_2']=l2
x['Description_3']=l3
x['Description_4']=l4
x['Description_5']=l5
```

```python
l1=[]
for i in x['Description_5']:
    try:
        temp=i.replace("\u200c","")
        l1.append(temp)
    except:
        l1.append(np.nan)
```
