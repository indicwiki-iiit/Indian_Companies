```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
# from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import numpy as np

s=Service("D:\chromedrive\chromedriver.exe")
driver=webdriver.Chrome(service=s)

df=pd.read_csv("final_wiki_1.csv")
l=[]
for link in df["url"]:
    try:
        driver.get(link)
        table= driver.find_elements(by = By.XPATH,value="//a[@title='More information about this page']")
        l.append(table[0].get_attribute("href"))
    except:
        l.append(np.nan)
l
```

```python
import numpy as np
s=Service("D:\chromedrive\chromedriver.exe")
driver=webdriver.Chrome(service=s)
temp=[]
ctr=0
for link in l:
    ctr+=1
    try:
        driver.get(link)
        table= driver.find_elements(by = By.XPATH,value="//table[@class='wikitable mw-page-info']")
        t=table[0].find_elements(by = By.XPATH,value="tbody/tr/td/a[@class='extiw wb-entity-link external']")
        temp.append(t[0].text)
    except:
        temp.append(np.nan)
    print(ctr)
```
